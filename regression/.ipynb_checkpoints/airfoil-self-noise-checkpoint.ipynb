```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

Attribute Information:

This problem has the following inputs: 
1. Frequency, in Hertzs. 
2. Angle of attack, in degrees. 
3. Chord length, in meters. 
4. Free-stream velocity, in meters per second. 
5. Suction side displacement thickness, in meters. 

The only output is: 
6. Scaled sound pressure level, in decibels. 

First 5 columns are the input, 6th column is the output

```python
data = pd.read_csv('airfoil_self_noise.dat', header=None, sep='\t')
data.head(10)
```

```python
data.describe()
#We need to scale the data, so big numbers don't dominate
```

```python
#Like to see the correlation between attributes

sns.pairplot(data)
```

```python
#Separate independent variables and dependent variables

X = data.iloc[:,0:-1]
y = data.iloc[:,-1]
```

```python
#Split data into training set and test set
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
#Scale independent variables

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

<h3>Random forest regression</h3>

```python
from sklearn.ensemble import RandomForestRegressor

test_scores=dict()

def random_forest(max_depth=3):
    regressor = RandomForestRegressor()    
    regressor.fit(X_train, y_train)
    print('Random forest score training with depth= ',max_depth,':',regressor.score(X_train, y_train))
    print('Random forest score testing  with depth= ',max_depth,':',regressor.score(X_test, y_test))
    test_scores[str(max_depth)] = regressor.score(X_test, y_test)
        

for md in range(100,210,10):
    random_forest(md)

print('-----------------------------------------------------------')

best_score = max(test_scores.values())
print(best_score)

best_depth = 100

for key, value in test_scores.items():
    if value == best_score:
        best_depth = value
        print('Best depth is: ', key)
        break
```
